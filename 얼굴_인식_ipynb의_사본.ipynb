- 얼굴이 학습된 파일 다운로드

```
!wget https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/haarcascade_frontalface_default.xml
!wget https://t1.daumcdn.net/cfile/tistory/99AE294C5C81F9DC16
!wget https://dimg.donga.com/wps/NEWS/IMAGE/2021/05/30/107175690.2.jpg
```

- 다운로드한 파일 불러옴

```
# imagePath = "99AE294C5C81F9DC16"
imagePath = "107175690.2.jpg"
cascPath = "haarcascade_frontalface_default.xml"

origin_image = cv2.imread(imagePath)
cv2_imshow(origin_image)
```

- 얼굴 인식률을 높히기 위해 사진을 회색조로 변경

```
image = cv2.imread(imagePath)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

- 사진에서 얼굴 감지

```
# Detect faces in the image
faceCascade = cv2.CascadeClassifier(cascPath)
faces = faceCascade.detectMultiScale(
   gray,
   scaleFactor=1.1,
   minNeighbors=5,
   minSize=(30, 30),
   flags = cv2.CASCADE_SCALE_IMAGE
)

print("Found {0} faces!".format(len(faces)))
```

- 감지된 얼굴에 초록색 박스 그리기

```
# Draw a rectangle around the faces
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)
```
